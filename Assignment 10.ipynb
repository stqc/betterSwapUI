## Finding out all null values in the data

```python
nanVals={}
for i in data.keys():
    nanVals[i]=len(data)-len(data[pd.notnull(data[i])])
```

```python
plt.bar(nanVals.keys(),nanVals.values())
plt.title("Bar Chart of all the items in data having null values")
plt.text(x=0,y=500,s="As we can see Cabin\nhas the most number of Null values\nfollowed by Age")
plt.show()
```
